```python
#now we merge for final, after fixing the timestamp issue.
```

```python
merged = pd.merge(
    trades,
    fear_greed,
    on='Date',
    how='inner'
)
```

```python
merged.shape
merged['classification'].value_counts()
#verification stage
```

```python
#so, the trader dataset itself contained mixed timestamp formats, like day-first string and many ambigious numerica fields.
#i validated the raw format, selected the explicit IST timestamp, applied the day-first parsing, cause panda assumes the format to be in MM-DD-YYYY, not like us
#this all to prevent silent misinterpretation and error.
```

```python
trades.columns = (
    trades.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
)
fear_greed.columns = fear_greed.columns.str.strip().str.lower()
```

```python
#now we will create a simpliefied sentiment column for ext.fear and ext.greed
merged['sentiment_group'] = merged['classification'].replace({
    'Extreme Fear': 'Fear',
    'Extreme Greed': 'Greed'
})
```

```python
merged['sentiment_group'].value_counts()
#jus checking
```

```python
merged.columns
```

```python
#we'll have to normalize them, Closed PnL exists.
merged.columns = (
    merged.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
)
```

```python
merged.groupby('sentiment_group')['closed_pnl'].mean().sort_values()
#re-run
```

```python
merged.groupby('sentiment_group')['size_usd'].mean()
```

```python
#interesting and totally expected, fear gets the traders moving in buying more, while the greed and neural as just momentum based.
```

```python
merged['pnl_per_usd'] = merged['closed_pnl'] / merged['size_usd']
merged.groupby('sentiment_group')['pnl_per_usd'].mean()
#profit efficiency
```

```python
import os

os.makedirs('outputs', exist_ok=True)
os.makedirs('csv_files', exist_ok=True)
```

```python
merged.to_csv(
    'csv_files/merged_trader_sentiment.csv',
    index=False
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(
    data=merged,
    x='sentiment_group',
    y='closed_pnl'
)
plt.title('Trader Profitability by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Closed PnL')

plt.savefig('outputs/pnl_by_sentiment.png', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=merged,
    x='sentiment_group',
    y='pnl_per_usd',
    estimator='mean'
)
plt.title('Profit Efficiency (PnL per USD) by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('PnL per USD')

plt.savefig('outputs/pnl_efficiency_by_sentiment.png', bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    data=merged,
    x='sentiment_group',
    y='size_usd',
    estimator='mean'
)
plt.title('Average Trade Size (USD) by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Size USD')

plt.savefig('outputs/size_usd_by_sentiment.png', bbox_inches='tight')
plt.show()
```
